```python
"""
F1 Round 13 — Hyperparameter-Tuned BBO Pipeline
==============================================
Function: Radiation Detection (2D)
Capstone Module 24 — Reinforcement Learning
"""

import json
import numpy as np
import pandas as pd
from datetime import datetime
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import Matern, RBF, ConstantKernel as C
from sklearn.svm import SVR
from sklearn.model_selection import LeaveOneOut
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from scipy.optimize import minimize
from scipy.stats import norm
from scipy.spatial.distance import cdist
import warnings
warnings.filterwarnings("ignore")

# ═════════════════════════════════════════════════════════════════════
# CONFIG
# ═════════════════════════════════════════════════════════════════════
FUNCTION_NAME    = "F1"
FUNCTION_LABEL   = "Radiation Detection"
N_DIMS           = 2
ROUND            = 13
DATA_DIR         = "function_1"

USE_SYMLOG       = True
LINTHRESH        = 1e-7
TIGHT_RADIUS     = 0.025
RANDOM_SEED      = 42

# ═════════════════════════════════════════════════════════════════════
# REUSABLE HELPERS
# ═════════════════════════════════════════════════════════════════════
def symlog10(y, linthresh=LINTHRESH):
    return np.sign(y) * np.log10(1.0 + np.abs(y) / linthresh)

def inv_symlog10(z, linthresh=LINTHRESH):
    return np.sign(z) * linthresh * (10.0 ** np.abs(z) - 1.0)

def to_raw(z):
    return inv_symlog10(z) if USE_SYMLOG else z

def make_kernel(kind, n_dims, ls_init=0.15, ls_bounds=(1e-4, 5.0)):
    base = C(1.0, (1e-4, 1e4))
    if kind.startswith("Matern_nu"):
        nu = float(kind.split("_nu")[1])
        return base * Matern(length_scale=[ls_init] * n_dims,
                             length_scale_bounds=ls_bounds, nu=nu)
    if kind == "RBF":
        return base * RBF(length_scale=[ls_init] * n_dims,
                          length_scale_bounds=ls_bounds)
    raise ValueError(f"Unknown kernel: {kind}")

def tune_gp(X, y_t, n_dims, kernels, alphas, n_restarts=15):
    rows = []
    for kname in kernels:
        for a in alphas:
            kernel = make_kernel(kname, n_dims=n_dims)
            gp = GaussianProcessRegressor(
                kernel=kernel, alpha=a,
                n_restarts_optimizer=n_restarts, normalize_y=True
            )
            gp.fit(X, y_t)
            lml = gp.log_marginal_likelihood(gp.kernel_.theta)
            rows.append({"kernel": kname, "alpha": a, "lml": float(lml),
                         "fitted": str(gp.kernel_)})
    return rows

def tune_svr(X_scaled, y_t, c_grid, gamma_grid, eps_grid):
    loo = LeaveOneOut()
    rows = []
    for c_val in c_grid:
        for g in gamma_grid:
            for eps in eps_grid:
                preds = np.zeros_like(y_t)
                for tr, te in loo.split(X_scaled):
                    m = SVR(kernel='rbf', C=c_val, gamma=g, epsilon=eps)
                    m.fit(X_scaled[tr], y_t[tr])
                    preds[te] = m.predict(X_scaled[te])
                mse = mean_squared_error(y_t, preds)
                rows.append({"C": c_val, "gamma": str(g), "eps": eps,
                             "loo_mse": float(mse)})
    return rows

def neg_ei(x, gp, y_best, xi=0.0):
    x = np.atleast_2d(x)
    mu, sigma = gp.predict(x, return_std=True)
    if sigma[0] < 1e-12:
        return 0.0
    z = (mu[0] - y_best - xi) / sigma[0]
    ei = (mu[0] - y_best - xi) * norm.cdf(z) + sigma[0] * norm.pdf(z)
    return -ei

def get_length_scales(gp):
    inner = gp.kernel_.k2
    return np.atleast_1d(inner.length_scale)

# ═════════════════════════════════════════════════════════════════════
# 1. LOAD DATA + SUMMARY
# ═════════════════════════════════════════════════════════════════════
X = np.load(f"{DATA_DIR}/initial_inputs.npy")
y = np.load(f"{DATA_DIR}/initial_outputs.npy")

assert X.shape[1] == N_DIMS, f"X has {X.shape[1]} dims, expected {N_DIMS}"

print("=" * 74)
print(f"FUNCTION {FUNCTION_NAME} — Round {ROUND}  ({FUNCTION_LABEL}, {N_DIMS}D)")
print("=" * 74)
print(f"Data points: {X.shape[0]}")
print(f"y range:     [{y.min():+.4e}, {y.max():+.4e}]")
print(f"Best point:  X={X[y.argmax()]}  y={y.max():+.4e}")
print(f"Worst point: X={X[y.argmin()]}  y={y.min():+.4e}")
print()

print("Top 5 by y:")
top5 = []
for i, idx in enumerate(np.argsort(y)[::-1][:5]):
    x_str = "[" + ", ".join(f"{v:.4f}" for v in X[idx]) + "]"
    print(f"  #{i+1} | y={y[idx]:+.4e} | X={x_str}")
    top5.append({"rank": i+1, "y": float(y[idx]), "x": X[idx].tolist()})
print("Bottom 3 by y:")
bot3 = []
for i, idx in enumerate(np.argsort(y)[:3]):
    x_str = "[" + ", ".join(f"{v:.4f}" for v in X[idx]) + "]"
    print(f"  #{i+1} | y={y[idx]:+.4e} | X={x_str}")
    bot3.append({"rank": i+1, "y": float(y[idx]), "x": X[idx].tolist()})
print()

x_best = X[y.argmax()].copy()
y_best_raw = float(y.max())

# ═════════════════════════════════════════════════════════════════════
# 2. PREPROCESSING
# ═════════════════════════════════════════════════════════════════════
print("─" * 74)
if USE_SYMLOG:
    print(f"PREPROCESSING: symlog10  (linthresh = {LINTHRESH:.0e})")
    print("─" * 74)
    y_t = symlog10(y)
    print(f"  Raw y:    min = {y.min():+.4e}   max = {y.max():+.4e}")
    print(f"  Symlog y: min = {y_t.min():+.4f}   max = {y_t.max():+.4f}")
    print(f"  Round-trip sanity: {np.allclose(inv_symlog10(y_t), y, rtol=1e-6)}")
else:
    print("PREPROCESSING: none (raw y values used)")
    print("─" * 74)
    y_t = y.copy()
print()
y_best_t = float(y_t[y.argmax()])

# ═════════════════════════════════════════════════════════════════════
# 3. GP TUNING
# ═════════════════════════════════════════════════════════════════════
print("=" * 74)
print("GP TUNING — kernel × alpha grid (metric: log marginal likelihood)")
print("=" * 74)

KERNELS = ["Matern_nu0.5", "Matern_nu1.5", "Matern_nu2.5", "RBF"]
ALPHAS  = [1e-8, 1e-6, 1e-4, 1e-2, 1e-1]

print(f"  {'kernel':14s} {'alpha':>7s} {'LML':>10s}   fitted")
print("  " + "-" * 100)
gp_results = tune_gp(X, y_t, N_DIMS, KERNELS, ALPHAS, n_restarts=15)
for r in gp_results:
    print(f"  {r['kernel']:14s} {r['alpha']:.0e} {r['lml']:>10.3f}   {r['fitted']}")

best_gp = max(gp_results, key=lambda r: r["lml"])
print(f"\n  ★ WINNER: {best_gp['kernel']} | alpha = {best_gp['alpha']:.0e} | "
      f"LML = {best_gp['lml']:.3f}")

gp_final = GaussianProcessRegressor(
    kernel=make_kernel(best_gp['kernel'], n_dims=N_DIMS),
    alpha=best_gp['alpha'], n_restarts_optimizer=40, normalize_y=True
)
gp_final.fit(X, y_t)
ls = get_length_scales(gp_final)
print(f"  Refitted kernel: {gp_final.kernel_}")
print(f"  Per-dim length scales (with 40 restarts):")
for d, l in enumerate(ls):
    flag = ""
    if l > 2.0:    flag = " ← very long (dim mostly irrelevant)"
    elif l > 0.5:  flag = " ← long (broad effect)"
    elif l < 0.05: flag = " ← short (highly sensitive)"
    print(f"    x{d+1}: {l:.4f}{flag}")

ls_upper = 5.0
if any(abs(l - ls_upper) < 0.01 for l in ls):
    print(f"  ⚠ One or more length scales pinned at upper bound ({ls_upper}).")
print()

# ═════════════════════════════════════════════════════════════════════
# 4. SVR TUNING
# ═════════════════════════════════════════════════════════════════════
print("=" * 74)
print("SVR TUNING — C × gamma × epsilon grid (metric: LOO-MSE on transformed y)")
print("=" * 74)

scaler_X = StandardScaler().fit(X)
X_scaled = scaler_X.transform(X)

SVR_C       = [0.1, 1.0, 10.0]
SVR_GAMMA   = ['scale', 0.1, 1.0, 10.0]
SVR_EPSILON = [0.001, 0.01, 0.1]

svr_results = tune_svr(X_scaled, y_t, SVR_C, SVR_GAMMA, SVR_EPSILON)
print(f"  {'C':>6s} {'gamma':>7s} {'eps':>7s}   {'LOO-MSE':>10s}")
print("  " + "-" * 38)
for r in sorted(svr_results, key=lambda r: r["loo_mse"])[:10]:
    print(f"  {r['C']:>6.1f} {r['gamma']:>7s} {r['eps']:>7.3f}   {r['loo_mse']:>10.4f}")
print(f"  ... ({len(svr_results) - 10} more combinations not shown)")

best_svr = min(svr_results, key=lambda r: r["loo_mse"])
print(f"\n  ★ WINNER: C={best_svr['C']}, gamma={best_svr['gamma']}, "
      f"eps={best_svr['eps']} | LOO-MSE = {best_svr['loo_mse']:.4f}")

g_final_raw = best_svr['gamma']
g_final = float(g_final_raw) if g_final_raw not in ('scale', 'auto') else g_final_raw
svr_final = SVR(kernel='rbf', C=best_svr['C'], gamma=g_final, epsilon=best_svr['eps'])
svr_final.fit(X_scaled, y_t)
print()

# ═════════════════════════════════════════════════════════════════════
# 5. PER-DIM NUDGE TEST  (smaller step sizes for tight convergence regime)
# ═════════════════════════════════════════════════════════════════════
print("=" * 74)
print("PER-DIM NUDGE TEST  (tuned GP)")
print("=" * 74)
print(f"  Anchor: best at {x_best}, y = {y_best_raw:+.4e}")
print()

NUDGES = [-0.005, -0.003, -0.001, +0.001, +0.003, +0.005]
nudge_log = []
print(f"  {'dim':>4s} {'delta':>7s} | {'GP_pred':>11s}   {'SVR_pred':>11s}  arrow")
print("  " + "-" * 58)
for d in range(N_DIMS):
    for delta in NUDGES:
        test = x_best.copy()
        test[d] = np.clip(test[d] + delta, 0, 1)
        gp_t  = float(gp_final.predict(test.reshape(1, -1))[0])
        gp_r  = float(to_raw(gp_t))
        svr_t = float(svr_final.predict(scaler_X.transform(test.reshape(1, -1)))[0])
        svr_r = float(to_raw(svr_t))
        arrow = "↑" if gp_t > y_best_t else "↓"
        clipped = " [CLIPPED]" if test[d] != x_best[d] + delta else ""
        print(f"   x{d+1}  {delta:+.4f} | {gp_r:>+11.4e}   {svr_r:>+11.4e}    {arrow}{clipped}")
        nudge_log.append({"dim": d+1, "delta": delta,
                          "gp_pred": gp_r, "svr_pred": svr_r})
    print()

# ═════════════════════════════════════════════════════════════════════
# 6. EI OPTIMIZATION
# ═════════════════════════════════════════════════════════════════════
print("=" * 74)
print(f"EI OPTIMIZATION  (tuned GP, ξ=0, radius=±{TIGHT_RADIUS})")
print("=" * 74)

tight_bounds = np.array([
    [max(0, x_best[i] - TIGHT_RADIUS), min(1, x_best[i] + TIGHT_RADIUS)]
    for i in range(N_DIMS)
])
print(f"  Search bounds: {tight_bounds.tolist()}")

rng = np.random.RandomState(RANDOM_SEED)
starts = [x_best.copy()]
for _ in range(300):
    starts.append(np.clip(x_best + rng.uniform(-TIGHT_RADIUS, TIGHT_RADIUS, N_DIMS), 0, 1))

best_val, best_x_ei = np.inf, None
for x0 in starts:
    res = minimize(neg_ei, x0, args=(gp_final, y_best_t, 0.0),
                   method='L-BFGS-B', bounds=tight_bounds)
    if res.fun < best_val:
        best_val, best_x_ei = res.fun, res.x.copy()

mu_ei, sigma_ei = gp_final.predict(best_x_ei.reshape(1, -1), return_std=True)
print(f"  GP-EI recommendation:  X = {best_x_ei}")
print(f"    GP pred:   {to_raw(mu_ei[0]):+.4e}  (± {sigma_ei[0]:.4f} in transformed space)")
print()

# ═════════════════════════════════════════════════════════════════════
# 7. CANDIDATE COMPARISON  (R12 — continue x2 trajectory)
# 5 consecutive new bests via small x2 step-downs.
# R10→R11 (x2: 0.679→0.674): +21% gain. GP under-predicted slightly.
# ═════════════════════════════════════════════════════════════════════
print("=" * 74)
print("CANDIDATE COMPARISON")
print("=" * 74)
# R12 strategy: continue proven x2 recipe with two step sizes.
# Add an x1 probe — x1 has been static at 0.685218 since R10, untested.
# GP-EI included as model recommendation.
# - x2 step down (-0.005): proven recipe, conservative continuation
# - x2 step down (-0.010): bolder, tests if gradient has room
# - x1 step up (+0.005): fresh-dim probe, x1 untested at recent anchors
# - GP-EI (tuned): model-driven recommendation
candidates = {
    "GP-EI (tuned)":             best_x_ei,
    "x2 step down (-0.005)":     np.array([0.685218, 0.668930]),
    "x2 step down (-0.010)":     np.array([0.685218, 0.663930]),
    "x1 step up (+0.005)":       np.array([0.690218, 0.673930]),
}
print(f"  {'name':22s} {'X':30s} {'GP_pred':>12s} {'SVR_pred':>12s} {'min_d':>8s}")
print("  " + "-" * 90)
candidate_log = []
for name, cand in candidates.items():
    gp_t  = float(gp_final.predict(cand.reshape(1, -1))[0])
    svr_t = float(svr_final.predict(scaler_X.transform(cand.reshape(1, -1)))[0])
    min_d = float(cdist(cand.reshape(1, -1), X).min())
    x_str = "[" + ", ".join(f"{v:.4f}" for v in cand) + "]"
    print(f"  {name:22s} {x_str:30s} {to_raw(gp_t):>+12.4e} "
          f"{to_raw(svr_t):>+12.4e} {min_d:>8.4f}")
    candidate_log.append({
        "name": name, "x": cand.tolist(),
        "gp_pred": float(to_raw(gp_t)),
        "svr_pred": float(to_raw(svr_t)),
        "min_dist_to_data": min_d,
    })
print()
gp_at_best = float(gp_final.predict(x_best.reshape(1, -1))[0])
svr_at_best = float(svr_final.predict(scaler_X.transform(x_best.reshape(1, -1)))[0])
print(f"  Sanity check (NOT a submission option):")
print(f"    GP prediction at current best: {to_raw(gp_at_best):+.4e}  "
      f"(actual: {y_best_raw:+.4e}, error: {to_raw(gp_at_best) - y_best_raw:+.4e})")
print(f"    SVR prediction at current best: {to_raw(svr_at_best):+.4e}  "
      f"(actual: {y_best_raw:+.4e}, error: {to_raw(svr_at_best) - y_best_raw:+.4e})")
print()

# ═════════════════════════════════════════════════════════════════════
# 8. FINAL R12 SUBMISSION OPTIONS
# ═════════════════════════════════════════════════════════════════════
print("=" * 74)
print(f"R{ROUND} SUBMISSION OPTIONS")
print("=" * 74)
for name, cand in candidates.items():
    print(f"  {name:22s}  {'-'.join(f'{v:.6f}' for v in cand)}")
print()
```

Reinforcement Learning Function
